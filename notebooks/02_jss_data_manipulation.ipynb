```python
import sqlite3
import os
import pandas as pd # type: ignore
db_path = os.path.join("..", "db", "database.db")
conn = sqlite3.connect(db_path)
cursor = conn.cursor()
```

## Select

```python
query_select_courses = """
SELECT * FROM COURSES
ORDER BY COURSE_NAME ASC;
"""
```

```python
# option 1 - how to print output with fetch
cursor.execute(query_select_courses)

rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
# option 2 - how to print output with pandas
df_courses = pd.read_sql(query_select_courses, conn)
df_courses.head()
```

```python
query_select_students = """
SELECT * FROM STUDENTS;
"""

df_students = pd.read_sql(query_select_students, conn)
df_students.head()
```

```python
conn.commit()
```

## Alter

```python
query_alter_instructor = """
ALTER TABLE INSTRUCTORS
ADD COLUMN SALARY REAL;
"""

cursor.execute(query_alter_instructor)
conn.commit()
```

```python
update_salaries = [
    (1, 25000.00),
    (2, 27000.00),
    (3, 30000.00),
    (4, 28000.00),
    (5, 32000.00)
]

for instructor_id, salary in update_salaries:
    cursor.execute("""
        UPDATE INSTRUCTORS
        SET SALARY = ?
        WHERE INSTRUCTOR_ID = ?
    """, (salary, instructor_id))

conn.commit()
```

```python
query_select_instructor = """
SELECT * FROM INSTRUCTORS;
"""

cursor.execute(query_select_instructor)
df_instructors = pd.read_sql(query_select_instructor, conn)
df_instructors.head()
```

## AND, NULL, IN, NOT

## COUNT, SUM

```python
query_select_students_courses = '''
SELECT COURSE_ID, COUNT(STUDENT_ID) 
FROM STUDENTS_COURSES
GROUP BY COURSE_ID
HAVING COUNT(STUDENT_ID) > 1;
'''

cursor.execute(query_select_students_courses)
df_students_courses =  pd.read_sql(query_select_students_courses, conn)
df_students_courses.head()
```

```python
query_select_instructor = """
SELECT SUM(SALARY) AS 'SALARY_TOTAL'
FROM INSTRUCTORS;
"""

cursor.execute(query_select_instructor)
df_instructors = pd.read_sql(query_select_instructor, conn)
df_instructors.head()
```

```python
query_select_instructor = """
SELECT SUM(SALARY) AS 'SALARY_TOTAL', AVG(SALARY) AS 'AVERAGE_SALARY'
FROM INSTRUCTORS;
"""

cursor.execute(query_select_instructor)
df_instructors = pd.read_sql(query_select_instructor, conn)
df_instructors.head()
```

```python
query_select_instructor = """
SELECT SUM(SALARY) AS 'SALARY_TOTAL', AVG(SALARY) AS 'AVERAGE_SALARY', MIN(SALARY) AS 'MIN_SALARY', MAX(SALARY) AS 'MAX_SALARY'
FROM INSTRUCTORS;
"""

cursor.execute(query_select_instructor)
df_instructors = pd.read_sql(query_select_instructor, conn)
df_instructors.head()
```

```python
conn.commit()
```
